```python
import pandas as pd
import random
```

```python
# --- new zell ---
df_pokemon = pd.read_csv('../src/pokemon.csv', encoding="utf-8-sig")
df_pokemon
```

```python
def add_spaces(x):
    c = random.choice([0,1,2])
    return c*" " + x + c*" "
```

```python
import numpy as np
df_pokemon['Name'] = np.where(df_pokemon['Legendary'], df_pokemon['Name'] + " " + "Legandary", df_pokemon['Name'])
```

```python
df_pokemon['Type 2'] = df_pokemon['Type 2'].astype(str)
df_pokemon['Type 1'] = df_pokemon['Type 1'].apply(add_spaces).str.lower()
df_pokemon['Type 2'] = df_pokemon['Type 2'].apply(add_spaces).str.lower()
df_pokemon['Name'] = df_pokemon['Name'].str.lower()
df_pokemon['HP'] = "HP" + df_pokemon['HP'].astype(str)
df_pokemon['Attack'] = "AT" + df_pokemon['Attack'].astype(str)
```

```python
# suchen in texten
df_pokemon[df_pokemon['Name'].str.contains('pika')]
```

```python
df_pokemon = df_pokemon[['Name', 'Type 1', 'Type 2', 'HP', 'Attack']]
df_pokemon.to_csv('../src/pokemon_defekt.csv', encoding="utf-8")
```
