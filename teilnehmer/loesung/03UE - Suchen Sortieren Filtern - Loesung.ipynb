## Teilaufgabe 5
Finde alle **Legandary** Pokemon und sortiere sie **absteigend nach ihrem Score (Total)**.

> Hinweis: Die Spalte Legendary ist laut ``df_pokemon.info()`` vom Typ boolean.<br>
dh. die Spalte ist bereits eine Series mit <font color="green">**True/False**</font> werten. 

```python
# --- Lösung ---
df_pokemon[df_pokemon['Legendary']].sort_values('Total', ascending=False)

# --- Alternativ Lösung ---
df_sort = df_pokemon[df_pokemon['Legendary']]
df_sort.sort_values('Total', ascending=False)

# --- Alternativ Lösung ---
# Wenn auch nicht ganz so schön da überflüssiger Vergleich mit True
df_pokemon[df_pokemon['Legendary'] == True].sort_values('Total', ascending=False).head()
```

## Teilaufgabe 6 
Zeige alle Pokemon, deren Wert für die Spalte **Total** zwischen 200 und 250 liegt.

```python
# --- Lösung ---
df_pokemon[(250 > df_pokemon['Total']) & (df_pokemon['Total'] > 200)]

# --- Alternative Lösung ---
# Entspricht der Dokumentation nach genau der obigen Lösung
df_pokemon[df_pokemon['Total'].between(200, 250, inclusive=False)]
```

## Zusatzaufgabe 3: Verschachtelte Logik
Wähle alle Pokemon in einem DataFrame, die
* vom **Typ Legendary** sind und deren **Total Wert mindestens 650** beträgt, oder
* die **mindestens** einen **Total Wert von 700 haben**, wenn Sie **nicht vom Typ Legendary** sind

Sortiere diese anschließend nach dem **Total Wert absteigend.**

```python
# --- Lösung ---
# Wenn die Statements sehr lange werden empfiehlt sich das ganze in zwei Schritte zu teilen. 
df_select = df_pokemon[(((df_pokemon['Legendary']) & (df_pokemon['Total'] >= 650)) | (df_pokemon['Total'] >= 700))]
df_select.sort_values('Total', ascending=False)

# --- Alternativ Lösung (Alles in einer Zeile)---
df_select = df_pokemon[(((df_pokemon['Legendary']) & (df_pokemon['Total'] >= 650)) | (df_pokemon['Total'] >= 700))].sort_values('Total', ascending=False)

# --- Alternativ Lösung (In Einzelschritten) ---
s_legendary = ((df_pokemon['Legendary']) & (df_pokemon['Total'] >= 650))
s_total = (df_pokemon['Total'] >= 700)

df_pokemon[s_legendary | s_total].sort_values('Total', ascending=False)
```
