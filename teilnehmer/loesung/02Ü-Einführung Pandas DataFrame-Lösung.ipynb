# Teilaufgabe 4
Selektiere und lasse dir jeweils folgende Datensätze anzeigen:
* Index ist eine ungerade Zahl
* Nur die Spalten **first_name, last_name, gender**
* Die Kombination aus beiden

```python
# --- Lösung --- 
df_kunden.loc[[1,3,5,7]]
```

```python
# --- Lösung --- 
df_kunden.loc[[1,3,5,7]][['first_name', 'last_name', 'gender']]

# --- Alternative Lösung --- 
df_select = df_kunden.loc[[1,3,5,7]]
df_select[['first_name', 'last_name', 'gender']]

# --- Alternative Lösung --- 
df_kunden.loc[[1,3,5,7],['first_name', 'last_name', 'gender']]
```
